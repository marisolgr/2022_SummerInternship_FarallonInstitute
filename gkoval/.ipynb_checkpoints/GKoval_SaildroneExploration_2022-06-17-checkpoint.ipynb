### Import Libraries

```python
#import necessary packages
import numpy as np
import pandas as pd
import xarray as xr
import matplotlib.pyplot as plt 
import matplotlib.dates as mdates
import seaborn as sns
import datetime
import warnings 
warnings.simplefilter('ignore') 

#load map packages
from cartopy.mpl.ticker import LongitudeFormatter, LatitudeFormatter
import cartopy.feature as cfeature
import cartopy.crs as ccrs
from calendar import month_abbr
```

## Saildrone

### Read in One Saildrone File

```python
ddir = '~/shared/users/mgarciareyes/saildrone_data/'
fn = 'saildrone_tpos_sd1069_2019_JPLv5.0_8dy_20210613norep_20210613_20211116.nc'
```

```python
ds = xr.open_dataset(ddir+fn)
ds.close()
ds
```

### Plot Trajectory

```python
#resample the data by hour
ds = ds.resample(time="H").mean(skipna=True)

#add date column but store it as a matplotlib date format
ds['date'] = mdates.date2num(ds['time'].dt.date)
```

```python
#define latitude and longitude boundaries
latr = [min(ds['lat']), max(ds['lat'])] 
lonr = [max(ds['lon']), min(ds['lon'])] 

# Select a region of our data, giving it a margin
margin = 0.5 
region = np.array([[latr[0]-margin,latr[1]+margin],[lonr[0]+margin,lonr[1]-margin]]) 

# Create and set the figure context
fig = plt.figure(figsize=(16,10), dpi = 72) 
ax = plt.axes(projection=ccrs.PlateCarree()) 
ax.coastlines(resolution='10m',linewidth=1,color='black') 
ax.add_feature(cfeature.LAND, color='grey', alpha=0.3)
ax.add_feature(cfeature.BORDERS)
ax.set_extent([region[1,0],region[1,1],region[0,0],region[0,1]],crs=ccrs.PlateCarree()) 
ax.set_xticks(np.round([*np.arange(region[1,1],region[1,0]+1,2)][::-1],0), crs=ccrs.PlateCarree()) 
ax.set_yticks(np.round([*np.arange(np.floor(region[0,0]),region[0,1]+1,1.5)],1), crs=ccrs.PlateCarree()) 
ax.xaxis.set_major_formatter(LongitudeFormatter(zero_direction_label=True))
ax.yaxis.set_major_formatter(LatitudeFormatter())
ax.gridlines(linestyle = '--', linewidth = 0.5)

# Plot track data, color by temperature
loc = mdates.AutoDateLocator()
sc = plt.scatter(x = ds['lon'], y = ds['lat'], c = ds['date'], cmap='viridis')
clb = fig.colorbar(sc, ticks=loc, format=mdates.AutoDateFormatter(loc))
clb.ax.set_title('Date')
plt.title('Sampling Track for Cruise', fontdict = {'fontsize' : 16})
plt.show()
```

### Plot One Variable

```python
#resample the data by Day
resampled_temp = ds.resample(time="D", skipna = True).mean()

#set figure size
plt.figure(figsize = (14,4))

#create lineplot 
sns.lineplot(x = ds['time'].dt.date, y = ds['TEMP_CTD_MEAN'], color = 'grey', ci = 'sd', alpha=0.2)
plt.scatter(resampled_temp['time'].dt.date, resampled_temp['TEMP_CTD_MEAN'], 30, [*range(len(resampled_temp['TEMP_CTD_MEAN']))], 'o', cmap='viridis')
plt.tight_layout()

#set labels
plt.title('Water Temperature for Saildrone Cruise', fontdict = {'fontsize' : 16})
plt.xlabel("Date")
plt.ylabel("Temperature [C]")

#reformat x-axis ticks
plt.grid(zorder=0, alpha=0.5)
plt.show()
```

### Plot Two Variables

```python
#visualize temp and salinity data with date
fig = plt.figure(figsize=(7.6,4)) 
loc = mdates.AutoDateLocator()
sc = plt.scatter(x = ds['TEMP_AIR_MEAN'], y = ds['TEMP_CTD_MEAN'], c = ds['date'], cmap='viridis', alpha = 0.5)
clb = fig.colorbar(sc, ticks=loc, format=mdates.AutoDateFormatter(loc))
#ax.set_label('Date')
plt.title('Air versus Water Temperature with Date', fontdict = {'fontsize' : 16})
plt.xlabel("Air Temperature [C]")
plt.ylabel("Water Temperature [C]")
plt.grid()
plt.show()
```

## Satellite

### Get Satellite Data

```python
region
```

```python
# input parameters

# select either a range of lat/lon or a point. 
latr = [-1.83678265, 23.32444742] 
lonr = [ -159.122864, -139.17382038] 

# time range. data range available: 2002-06-01 to 2020-01-20. [start with a short period]
dater = ['2019-06-08','2019-06-28'] # dates on the format 'YYYY-MM-DD' as string

# first determine the file name using, in the format:
# the s3 bucket [mur-sst], and the region [us-west-2], and the folder if applicable [zarr-v1] 
file_location = 'https://mur-sst.s3.us-west-2.amazonaws.com/zarr-v1'

ds_sst = xr.open_zarr(file_location,consolidated=True) # open a zarr file using xarray
# it is similar to open_dataset but it only reads the metadata

ds_sst 
```

```python
#remove all values that are for lakes (look at the meta data for the mask field above)
sst_filtered = ds_sst.where(ds_sst.mask != 5, np.nan)

#filter the data using the specified extent of the latitutde, longitude, and time from above
sst = sst_filtered['analysed_sst'].sel(time = slice(dater[0],dater[1]),
                                       lat  = slice(latr[0], latr[1]), 
                                       lon  = slice(lonr[0], lonr[1]))

sst = sst-273.15 # transform units from Kelvin to Celsius
sst.attrs['units']='deg C' # update units in metadata
```

## Plot Satellite Data

```python
#define latitude and longitude boundaries
latr = [min(sst['lat']), max(sst['lat'])] 
lonr = [max(sst['lon']), min(sst['lon'])] 

# Select a region of our data, giving it a margin
margin = 0.5 
region = np.array([[latr[0]-margin,latr[1]+margin],[lonr[0]+margin,lonr[1]-margin]]) 

# Create and set the figure context
fig = plt.figure(figsize=(16,10), dpi = 72) 
ax = plt.axes(projection=ccrs.PlateCarree()) 
ax.coastlines(resolution='10m',linewidth=1,color='black') 
ax.add_feature(cfeature.LAND, color='white', alpha=0.3)
ax.add_feature(cfeature.BORDERS)
ax.set_extent([region[1,0],region[1,1],region[0,0],region[0,1]],crs=ccrs.PlateCarree()) 
ax.set_xticks(np.round([*np.arange(region[1,1],region[1,0]+1,2)][::-1],0), crs=ccrs.PlateCarree()) 
ax.set_yticks(np.round([*np.arange(np.floor(region[0,0]),region[0,1]+1,5)],1), crs=ccrs.PlateCarree()) 
ax.xaxis.set_major_formatter(LongitudeFormatter(zero_direction_label=True))
ax.yaxis.set_major_formatter(LatitudeFormatter())
#ax.gridlines(linestyle = '--', linewidth = 0.5)

# Plot track data, color by temperature
sst.mean(dim={'time'}, skipna=True, keep_attrs=True).plot(transform=ccrs.PlateCarree(),cbar_kwargs={'label': 'Temperature [deg C]'}, cmap = "coolwarm")
plt.scatter(x = ds['lon'], y = ds['lat'], c = 'darkgrey')
plt.title('Sampling Track for Cruise with Averaged MUR SST Data', fontdict = {'fontsize' : 16})
plt.show()
```
