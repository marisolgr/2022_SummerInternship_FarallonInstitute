```python
import numpy as np
import pandas as pd
import xarray as xr
import matplotlib.pyplot as plt
import math
import glob
import cartopy.crs as ccrs
from cartopy.mpl.ticker import LongitudeFormatter, LatitudeFormatter
import cartopy.feature as cfeature
import warnings
warnings.simplefilter('ignore') # filter some warning messages
```

```python
ddir = '../saildrone data/'
fns = glob.glob(ddir+'*.nc')
#fns
```

```python
plt.figure(figsize=(12,6))
d = 0
for i in fns:
    ds = xr.open_dataset(i)
    ds.close()  
    # map
    plt.scatter(ds.time,ds['lat']*0+d, 10,ds.time.dt.month,'o',cmap='jet', vmin=1, vmax=12)
    d +=1
    
plt.colorbar()
plt.grid(zorder=0, alpha=0.5)
plt.show()
```

```python
d = 0
plt.figure(figsize=(12,6))
for i in fns:
    ds = xr.open_dataset(i)
    ds.close()  
    # map
    plt.scatter(ds.time.dt.month,ds['lat']*0+d, 40,ds.lat,'o',cmap='RdYlBu_r', vmin=-5, vmax=80, alpha=0.5)
    d +=1
    
cbar = plt.colorbar()
cbar.set_label('Latitude')
plt.xlabel('Month')
plt.ylabel('Cruise')
plt.grid(zorder=0, alpha=0.5)
plt.show()
```
